```python
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
mpg
```

## Create a visual showing the relationship between horsepower and fuel economy.


```python
sns.relplot(data=mpg
            ,x='horsepower'
            ,y='mpg')
# The higher the mpg, the lower the horsepower
```

## Create a table summarizing fuel economy by country of origin


```python
# Get the average mpg by country of origin
avg_mpg_by_origin = mpg.groupby('origin')['mpg'].mean().reset_index()
```

```python
# Sort values
avg_mpg_by_origin = avg_mpg_by_origin.sort_values(by='mpg', ascending=False)
print(avg_mpg_by_origin)
# Japan has the highest average fuel economy compared to Europe and USA
```

## Create a visual showing how fuel economy has changed over time


```python
# Get the average mpg by year, easier to read by year
avg_mpg_by_year = mpg.groupby('model_year')['mpg'].mean().reset_index()
sns.lineplot(data=avg_mpg_by_year, x='model_year', y='mpg', marker='o')
# The fuel economy, mpg, increased over time
```

## Create two other visuals or tables related to fuel economy.

```python
# Visual to show which cylinder offers better fuel economy 
sns.barplot(data=mpg, x='cylinders', y='mpg', errorbar=None)
# On average, vehicles with 4 cylinders has the best fuel economy followed by 5 cylinders
```

```python
# Visual to show the relationship between horsepower, fuel economy, and weight
sns.relplot(data=mpg
            ,x='horsepower'
            ,y='mpg'
           ,hue='weight')
# As the vehicle becomes heavier, the horsepower increases, but the fuel economy decreases.
```
